#### Ejercicio de funciones

- Una funcion que reciba el numero de coches de un concesionario y devuelva un diccionario con matricula como clave y una lista con modelo, color, km e itv como valor
- Una funcion que recibe ese diccionario y una itv y devuelve una lista con las matriculas de los coches que han de pasar la itv ese aÃ±o
- Una funcion que recibe el diccionario anterior y devuelve una lista con la matricula/matriculas de los coches con  mayor km totales recorridos.

```python
def llenar_diccionario(num_coches:int) -> dict :
    coches = {}
    for i in range(num_coches):
        mat = input("Introduce la matricula: ")
        while mat in coches:
            mat = input("Introduce la matricula: ")
            
        modelo = input("Modelo: ")
        color = input("Color: ")
        km_str = input("Km totales: ")
        km_int = int(km_str)
        itv_str = input("Itv: ")
        itv_int = int(itv_str)
        coches[mat] = [modelo, color, km_int, itv_int]
    return coches
        
    
    
    
def checkear_itv(coches: dict, itv:int)-> list:
    matriculas = []
    for mat in coches:
        valor = coches[mat]
        itv_coche = valor[3]
        #itv_coche = coches[mat][3]
        if itv_coche == itv:
            matriculas.append(mat)
            
    return matriculas
        
        
def obtener_maximos(coches:dict)-> list:
    matriculas = []
    maximo = 0
    ### Hacer una funcion que solo obtenga el maximo y lo devuelva
    for mat in coches:
        valor = coches[mat]
        km_totales = valor[2]
        #km_totales = coches[mat][2]
        if km_totales >= maximo:
            maximo = km_totales
       
       
    ### hacer una funcion que solo haga estoo     
    for mat in coches:
        if coches[mat][2] == maximo:
            matriculas.append(mat)
            
    return matriculas
        
    
        
        
        
```

```python
### Programa principal

# coches = llenar_diccionario(3)

coches = {'1234H': ['Golf', 'rojo', 10000, 2025],
          '1672K': ['Seat', 'Negro', 120000, 2025], 
          '12312k': ['Mercedes', 'Verde', 120000, 2024]}

matriculas = obtener_maximos(coches)
print(matriculas)
```

```python
def crear_tablero(nfil:int, ncol:int)-> list:
    tablero_funcion = []
    for i in range(nfil):
        fila = []
        for j in range(ncol):
            fila.append("_")
        tablero_funcion.append(fila)
    return tablero_funcion


def pintar_tablero(tablero:list)-> None:
    for fila in tablero:
        for elem in fila:
            print(elem, end = " ")
        print()
            
    
```

```python
filas = 4
columnas = 4
tablero = crear_tablero(filas,columnas)

pintar_tablero(tablero)
```
