```python
class SumandoNoValidoError(Exception):
    pass
```

```python
def sumar(a:str, b:str) -> int:
    """
    Realiza la suma de operandos positivos comprendidos en el rango [1, 100]
    """
    try:
        a_int = int(a)
        b_int = int(b)
        if a_int > 100 or a_int < 1:
            raise SumandoNoValidoError(f"Sumando \"{a_int}\" no válido")    
        if b_int > 100 or b_int < 1:
            raise SumandoNoValidoError(f"Sumando \"{b_int}\" no válido")  
        return a_int + b_int
    except ValueError as error:
        mensaje = error.args[0].lstrip("invalid literal for int() with base 10: ")
        mensaje = mensaje.replace("\'", "")
        raise SumandoNoValidoError("Error en el cambio de tipo")

try:
    suma = sumar("9a", "400")
    print(f"El resultado de la suma es {suma}")
except SumandoNoValidoError as error:
    print(error)
```

```python
def crear_persona(nombre:str, dni:str) -> tuple:
    """
    Crea una tupla representando a una persona, usando su nombre y DNI.

    Esta función toma como parámetros el nombre y el DNI de una persona y 
    devuelve una tupla que contiene ambos datos.

    Args:
        nombre (str): El nombre de la persona.
        dni (str): El Documento Nacional de Identidad de la persona con letra.

    Returns:
        tuple: Una tupla que contiene una persona: nombre y DNI.
    
    Raises:
        DNIError: Si la cadena dada no es un DNI con números + letra.

    Examples:
        >>> crear_persona("Juan Perez", "12345678A")
        ("Juan Perez", "12345678A")
    """
    if type(nombre) == str and type(dni) == str:
        persona = ()
        if not dni[-1:].isdecimal() and dni[:-1].isdecimal():
            persona = (nombre, dni)   
        else:
            raise DNIError("El DNI debe ser una secuencia de números seguido de una letra")  # Aquí el código se interrumpirá y no se seguirá ejecutando el programa

        return persona
    else:
        raise TypeError("Los argumentos deben ser str")

try:
    persona = crear_persona("Luis", "1111111")
    print(persona)
except DNIError as error:
    print("DNIError:", error)
except TypeError as error:
    print("Los tipos no coinciden:", error)
```
