```python
f = open("codificaciones.txt", "r", encoding = "utf-8")
separador = ";"
header = f.readline()
header_spliteado = header.split(";")

# PENSAR COMO ITERAR EL HEADER POST SPLITEO Y ASIGNAR
# POSICIONES DINÁMICAS
traductor = {}
for lenguaje in header_spliteado:
    traductor[lenguaje] = {}

for linea in f:
    linea = linea.strip()
    lista = linea.split(separador)
    latin = lista[0]
    morse = lista[1]
    radio = lista[2]
    traductor["Morse"][latin]  = morse
    traductor["Radio"][latin] = radio
f.close()
print(traductor)
```
