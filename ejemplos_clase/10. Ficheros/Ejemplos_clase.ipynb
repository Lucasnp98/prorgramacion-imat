```python
def escribir_fichero(dic_coches : dict, nombre_fichero: str) -> None:
    # ojo al encoding....xº
    f = open(nombre_fichero, "w",   encoding = "utf-8")

    header = "Matrícula,Modelo,Color,itv\n"
    f.write(header)
    for mat in dic_coches:
        v = dic_coches[mat]
        
        linea  = f"{mat},{v[0]},{v[1]},{v[2]}\n"
        
        f.write(linea)

    f.close()
    
    

        
```

```python
if __name__ == "__main__":
    coches = {"1234KCD":["Mercedes", "Negro", 2025],
          "1847JKB":["BMW", "Gris", 2024]}
    escribir_fichero(coches, "coches.csv")
    
```

```python
def leer_fichero(nombre: str)-> dict:
    f = open(nombre, "r")
    coches = {}
    header = f.readline()
    for linea in f:
        linea = linea.strip()
        lista = linea.split(",")
        mat = lista[0]
        modelo = lista[1]
        color = lista[2]
        itv = int(lista[3])
        coches[mat] = [modelo, color, itv]

    f.close()
    
    return coches


    
```

```python
if __name__ == "__main__":
    dic_coches = leer_fichero("coches.csv")
    print(dic_coches)
```
